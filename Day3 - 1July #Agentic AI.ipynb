```python
import requests
import json

URL = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
    "Content-Type": "application/json"
}

payload = {
    "model": MODEL_NAME,
    "messages": [
        {"role": "user", "content": "Why is the sky blue? Answer in exactly 2 sentences."}
    ],
    "temperature": 0.4,
    "stream": False # Get the full answer at once
}

print("Sending request to OpenRouter Cloud...\n")
response = requests.post(URL, headers=headers, data=json.dumps(payload))

if response.status_code == 200:
    data = response.json()
    # Extract response according to OpenAI/OpenRouter schema standard
    reply = data["choices"][0]["message"]["content"]
    print("🤖 AI Response:")
    print(reply)
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
import requests
import json

URL = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
    "Content-Type": "application/json"
}

payload = {
    "model": MODEL_NAME,
    "messages": [
        {"role": "user", "content": "Write a short poem about coding APIs in the cloud."}
    ],
    "stream": True # Enable streaming delivery
}

response = requests.post(URL, headers=headers, data=json.dumps(payload), stream=True)

print("🤖 Streaming AI Response:\n")
for line in response.iter_lines():
    if line:
        decoded_line = line.decode('utf-8').strip()

        # OpenRouter streams chunks prefixed with "data: "
        if decoded_line.startswith("data: "):
            content = decoded_line[6:]

            # Check for stream end signal
            if content == "[DONE]":
                break

            try:
                chunk_json = json.loads(content)
                delta = chunk_json['choices'][0]['delta']
                if 'content' in delta:
                    print(delta['content'], end="", flush=True)
            except Exception:
                pass
print()
```

```python
import requests
import json

URL = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
    "Content-Type": "application/json"
}

# Conversation memory array
messages = [
    {"role": "system", "content": "You are a helpful, concise assistant running inside a Google Colab environment."}
]

print("🤖 OpenRouter Chatbot Active! Type 'quit' or 'exit' to terminate.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ("quit", "exit", "bye"):
        print("👋 Chat session ended.")
        break

    if not user_input.strip():
        continue

    # 1. Append user input to history
    messages.append({"role": "user", "content": user_input})

    payload = {
        "model": MODEL_NAME,
        "messages": messages,
        "stream": False
    }

    try:
        response = requests.post(URL, headers=headers, data=json.dumps(payload))

        if response.status_code == 200:
            reply = response.json()["choices"][0]["message"]["content"]
            print(f"\nAI: {reply}\n")

            # 2. Append AI response back into context memory
            messages.append({"role": "assistant", "content": reply})
        else:
            print(f"\nAPI Error {response.status_code}: {response.text}")
            break

    except Exception as e:
        print(f"\nConnection Error: {e}")
        break
```
